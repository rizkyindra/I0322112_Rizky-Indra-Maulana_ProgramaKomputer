```python
#Buatlah sebuah fungsi untuk mencari invers matrix berdasarkan input matrix dari user

def invers():
    det = []

    #Tulis kode kamu disini
def invers():
    baris = int(input('Masukkan Banyak Baris: '))
    kolom = int(input('Masukkan Banyak Kolom: '))
    
    matrix = create_matrix(baris, kolom)
    det = determinan(matrix)

    if det == 0:
        print("Matriks tidak memiliki invers karena determinan sama dengan 0.")
    else:
        inverse_matrix = inverse(matrix, det)
        print("Hasil Pemrograman")
        print("Hasil Matriks invers:")
        print()
        print_matrix(inverse_matrix)


def create_matrix(baris, kolom):
    matrix = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            element = float(input(f'Inputkan elemen matriks pada baris ke-{i+1}, kolom ke-{j+1}: '))
            row.append(element)
        matrix.append(row)
    return matrix


def determinan(matrix):
    n = len(matrix)
    if n != len(matrix[0]):
        raise ValueError("Matriks diharuskan berbentuk persegi (jumlah baris = jumlah kolom).")
    
    if n == 1:
        return matrix[0][0]
    
    det = 0
    for y in range(n):
        submatrix = []
        for x in range(1, n):
            row = matrix[x][:y] + matrix[x][y+1:]
            submatrix.append(row)
        det += (-1) ** y * matrix[0][y] * determinan(submatrix)
    
    return det


def cofactor(matrix):
    n = len(matrix)
    cofactor_matrix = []
    for x in range(n):
        row = []
        for y in range(n):
            submatrix = []
            for k in range(n):
                if k != x:
                    subrow = matrix[k][:y] + matrix[k][y+1:]
                    submatrix.append(subrow)
            sign = (-1) ** (x + y)
            cofactor_ij = sign * determinan(submatrix)
            row.append(cofactor_ij)
        cofactor_matrix.append(row)
    return cofactor_matrix


def transpose(matrix):
    n = len(matrix)
    transposed_matrix = []
    for y in range(n):
        row = []
        for x in range(n):
            row.append(matrix[x][y])
        transposed_matrix.append(row)
    return transposed_matrix


def inverse(matrix, det):
    n = len(matrix)
    cofactor_matrix = cofactor(matrix)
    adjugate_matrix = transpose(cofactor_matrix)
    inverse_matrix = []
    for x in range(n):
        row = []
        for y in range(n):
            element = adjugate_matrix[x][y] / det
            row.append(element)
        inverse_matrix.append(row)
    return inverse_matrix


def print_matrix(matrix):
    for row in matrix:
        print(row)


invers()

print()
```
